```python
import pandas_datareader.data as web
import datetime
```

```python
symbol="AAPL"
day_start=datetime.datetime(2010,1,1)
day_end = datetime.datetime(2016,11,28)
```

```python
df3 = web.DataReader(symbol,"yahoo",day_start,day_end)
```

```python
df3.head()
```

```python
df3.Close.plot() #收盤價，預設折線圖
```

# Close 後ㄧ項減前ㄧ項

```python
df3["Diff"] = df3.Close.diff()
```

```python
df3.head()
```

```python
df3.Diff.plot()
```

```python
df3["Rel_Diff"]=df3.Diff / df3.Close
```

```python
df3.head()
```

```python
df3.dropna(inplace=True)
```

```python
df3.head()
```

```python
df3.Diff.min()
```

```python
df3.Diff.argmin()
```

```python
df3.drop(df3.Diff.argmin(),inplace=True)
```

```python
df3.Diff.plot()
```
