```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
from sklearn.cross_validation import train_test_split
```

```python
x = np.linspace(0,1,100)
y = 3*x**3 + 5*x + 0.5*np.random.randn(100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
def f(x,a,b):
    return a*x**3 + b*x
popt, pcov = curve_fit(f, x_train, y_train)
a,b = popt
print(a,b)
plt.scatter(x,y,c="#50C5B7",s=30);
plt.plot(x,3*x**3 + 5*x,lw=3,c='r');
plt.plot(np.sort(x_train),a*np.sort(x_train)**3 + b*np.sort(x_train),c='#AB8476');
plt.plot(np.sort(x_test),a*np.sort(x_test)**3 + b*np.sort(x_test),c='#F0F465');
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data # 平常的x，有4個參數 (花萼長度、寬度、花瓣長度、寬度) features
```

```python
iris.target # 平常的y，由0，1，2表示不同種的鳶尾花 分類label
```

```python
x = iris.data[:,:2]
y = iris.target
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

```python
grid = np.array([(i,j) for i in np.arange(3,8,0.3) for j in np.arange(1,7,0.4)])
```

```python
co = clf.predict(grid) 
```

```python
plt.scatter(grid[:,0],grid[:,1],c=co,s=50);
```
