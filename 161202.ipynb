```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv")
```

```python
df_continent=df.groupby("continent").mean()
```

```python
df_continent.plot(kind="bar");
plt.legend(bbox_to_anchor=(1.04,1),loc=2,borderaxespad=0.);
```

```python
df_continent.plot(kind="bar",subplots=True);
```

```python
df_continent.plot(kind="barh")
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
df_continent.plot(kind="bar")
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
df_continent.plot(kind="bar",subplots=True,layout=(1,4))
plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
```

```python
df_continent.plot(kind="bar",subplots=True,layout=(2,2));
```

```python
df_continent.plot(kind="line",subplots=True,layout=(2,2),legend= False);
```
